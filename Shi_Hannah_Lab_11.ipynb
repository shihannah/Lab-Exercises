```python
import random
from tkinter import *
import tkinter as tk
import requests

root = Tk()
root.title("Trivia")
root.geometry("800x900")


url = requests.get(f"https://the-trivia-api.com/api/questions?categories=film_and_tv&limit=10&region=US&difficulty=easy")
json = url.json()

canvas = Canvas(root, width=700, height=350, bg="darkseagreen")
canvas.place(relx=.5, rely=.5,anchor= CENTER)

game_display = canvas.create_text(200, 75, text="", fill="black", font=('arial 15 bold'))
feedback = Label(root, text="", pady=10, font=("arial", 15, "bold"))

user_answer = StringVar()
user_answer.set("")
score = 0
questions_correct = 0
        
def get_questions():
    global list_questions
    list_questions = []
    for i in json:
        question = i['question']
        list_questions.append(question)
    
def get_answers():
    global random_question
    random_question = random.choice(list_questions)
    
    global options
    options = []
    
    for i in json:
        if random_question == i['question']:
            global correct_ans
            correct_ans = i['correctAnswer']
            options.append(correct_ans)
            for incor in i["incorrectAnswers"]:
                options.append(incor)
            list_questions.remove(random_question)
                
    random.shuffle(options)
    
def display_question():
    canvas.itemconfig(game_display, text=random_question, width=300)
    
def radio_button():
    y_pos = 100
    for i in options:
        global rad_btn
        rad_btn = Radiobutton(canvas, text=i, variable=user_answer, command=score, value=i, font=('arial 10 bold'))
        rad_btn.place(x=50, y=y_pos)
        y_pos+=20        
        
def score(): #do not know how to get 'incorrect' to stop appearing before user answers question
    ans = user_answer.get()
    for i in json:
        if random_question == i['question']:
            if ans == correct_ans:
                feedback['text'] = 'CORRECT!'
                feedback.place(x=350,y=450)
                global score
                global questions_correct
                score+=10
                questions_correct+=1
            else:
                feedback['text'] = 'INCORRECT :('
                feedback.place(x=350,y=450)  
    return score
    return questions_correct
            
def completed():
    if list_questions == []:
        for i in options:
            canvas.itemconfig(game_display, text=f'You have answered all the questions! You have {score} points!', width=300) #score doesn't show up properly
            
def next_question():
    try:
        for i in options: #could not get questions to not overlap so some questions and options show behind the current questions and options
            rad_btn.destroy()
        get_answers()
        display_question()
        radio_button()
        score()
    except:
        completed()
    
def next_button():
    nextButton = Button(canvas, text="Next", font=("arial", 10), command=next_question)
    nextButton.place(x=600,y=250)
    
def game():
    get_questions()
    get_answers()
    display_question()
    radio_button()
    score()
    completed()
    next_question()
    next_button()
    
        
game()
root.mainloop() 
```
